# Introduction to Statistical Graphics

Using R, we can generate lots of interesting statistical graphics.

R has lots of built in datasets.

One of these is hair and eye color data.  These are examples of **qualitative** or **categorical** data \(i.e., not numbers\).


```r
HairEyeColor
```

```r
help(HairEyeColor)
```

```r
mosaicplot(~ Hair + Eye, data=HairEyeColor)
```

```r
mosaicplot(~ Hair + Sex, data=HairEyeColor)
```

```r
mosaicplot(~ Eye + Sex, data=HairEyeColor)
```

It appears that hair and eye color are dependent random variables, but sex is independent with both hair color and eye color.

Now let's switch to some measurements of the speed of light.  This is an example of **numerical** or **quantitative** data.


```r
speed.measurements <- scan("michelson1.d")
```

```r
speed.measurements
```

```r
sort(speed.measurements)
```

One way of visualizing numerical data is a **stem and leaf plot**.


```r
stem(speed.measurements)
```

Another similar way of visualizing this data set is a **histogram**.


```r
help(hist)
```

```r
hist(speed.measurements)
```

```r
sort(speed.measurements)
```

We can also use a **boxplot**.  This highlights the **quartiles** of the data.


```r
boxplot(speed.measurements)
```

```r
quantile(speed.measurements, c(0.25, 0.5, 0.75))
```
