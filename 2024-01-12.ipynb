To get a subset of a datset, we can use the **subset** function.


```r
setosa <- subset(iris, Species == "setosa")
setosa
```

The following gives an **empirical cumulative distribution plot**.  It gives the proportion of specimens with a petal length less than or equal to the given x\-value.


```r
plot.ecdf(setosa$Petal.Length)
```

When a data set has multiple levels like iris does, we can create multiple side\-by\-side box plots.


```r
boxplot(Petal.Length ~ Species, data=iris)
```

We've used the tilde (`~`) operator a few times.  What it does is create formulas.  In general, we use `y ~ x`.  The dependent variable (or in statistics terms, the **response** variable) goes on the left and the independent variable (or **explanatory** variable) goes on the right.

Using this, we can create **scatter plots**.  Each point in a scatter plot is of the form $(x_i, y_i)$, where $x_i$ is the value of the explanatory variable for the $i$th element of the data set and $y_i$ is the value of the response variable for the $i$th element of the data set.

So for example, suppose we want to visualize the relationship between petal length and sepal length of iris setosa flowers.


```r
plot(Sepal.Length ~ Petal.Length, data=setosa)
```

It appears that the longer the petal, the longer the sepal, which makes intuitive sense!
