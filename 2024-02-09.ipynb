# Maximum Likelihood Estimators

Recall that if $X_1,\ldots,X_n\sim\operatorname{Exp}(\lambda)$, we found that the MLE (maximum likelihood estimator) of $\lambda$ is $\hat\lambda=\frac{1}{\overline X}$.

Let's check this with R:

```r
x <- rexp(1000, 8)
x
```

```r
lambdaHat <- 1/mean(x)
lambdaHat
```

Not bad!

Now, let's plot the empirical cumulative distribution function together with the cumulative distribution function of an $\operatorname{Exp}(\hat\lambda)$ random variable.

```r
plot.ecdf(x)
xx <- seq(0, 0.8, 0.05)
yy <- pexp(xx, lambdaHat)
lines(xx, yy, col="red")
```
